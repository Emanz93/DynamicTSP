```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
loaded = np.load('../instances/iridium_weights_matricies.npz')
weights = loaded['w']
loaded.close()
N_CITIES = np.shape(weights)[1]
```

```python
G.add_nodes_from(list(range(N_CITIES)))
G.number_of_nodes()
```

```python
for i in range(N_CITIES):
    for j in range(i + 1, N_CITIES):
        G.add_weighted_edges_from([(i,j,weights[0, i, j])])
        G.add_weighted_edges_from([(j,i,weights[0, i, j])])
G.number_of_edges()
```

```python
G[1][2]['weight']
```

```python
# Fruchterman Reingold
fr_positions = nx.fruchterman_reingold_layout(G)
nx.draw(G, node_size=20, node_color="skyblue", pos=fr_positions)
plt.title("fruchterman_reingold")
```

```python
# Spring
s_position = nx.spring_layout(G)
nx.draw(G, node_size=20, node_color="skyblue", pos=s_position)
plt.title("spring")
```
